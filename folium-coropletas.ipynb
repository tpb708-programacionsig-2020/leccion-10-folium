## Clase [Choropleth](https://python-visualization.github.io/folium/modules.html#folium.features.Choropleth)

Crea un mapa de coropletas como una capa sobrepuesta sobre el mapa base y la colorea de acuerdo al valor de un campo de datos.

```python
df = pd.read_csv('datos/10_07_CSV_POSITIVOS_ULTIMA_FECHA.csv', sep=',')
```

```python
df.rename(columns = {'Unnamed: 0': 'id'}, inplace=True)
```

```python
df.head(5)
```

```python
# file name - file is located in the working directory
geo = r'datos/cr_limite_cantonal_ign_wgs84.geojson' # geojson file

# create a plain world map
map = folium.Map(location=[10, -84], zoom_start=6, tiles='stamenwatercolor')


# generate choropleth map 
map.choropleth(
    geo_data=geo,
    data=df,
    columns=['cod_canton', 'positivos'],
    key_on='feature.properties.cod_canton',
    fill_color='YlGnBu', 
    fill_opacity=1, 
    line_opacity=1,
    legend_name='Casos positivos',
    smooth_factor=0)


map
```
